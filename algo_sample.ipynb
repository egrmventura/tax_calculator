```python
test = [3, 5, 6, 7, 6, 5, 5, 3, 6]

def boyer_moore_multi(list):
    leads_list = boyer_moore_indiv(list)
    max_count = leads_list[1]
    leads = []
    next_count = max_count
    while next_count == max_count:
        leads.appdend(leads_list[0])
        leads_list = boyer_moore_indiv(list, leads)
        next_count = leads_list[1]

    return leads
    
def boyer_moore_indiv(list, last_leads=None):
    count = 0
    lead = None
    for n in list:
        if count == 0:
            if last_leads != None and n in last_leads:
                count-=1
            else:
                lead = n
                count+=1
        else:
            count += 1 if lead == n else -1
    return [lead, count]

print(boyer_moore_multi(test))
```
